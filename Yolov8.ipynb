```python
# TRAIN YOLO
from ultralytics import YOLO

# Load a pretrained YOLO model
model = YOLO("yolov8l.pt")

# Train the model
model.train(data="/content/drive/MyDrive/Colab Notebooks/Leishmania-Data_RAW/LEISHMANIA_NEW_LABEL/Leishmania-Data_RAW/Labels/YOLODataset/dataset.yaml", epochs=50)

# Evaluate model performance on the validation set
metrics = model.val()

# Print evaluation metrics
print(metrics)
print("--------------------------------------------------")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img_path = ["deneme.jpg","deneme2.jpg","deneme3.jpg","deneme4.jpg","deneme5.jpeg"]

results = model(img_path)

for i, r in enumerate(results):
 im_bgr = r.plot()
 im_rgb = Image.fromarray(im_bgr[..., ::-1])
 plt.figure(figsize=(8, 6))
 plt.imshow(im_rgb)
 plt.title(f'Result {i+1}')
 plt.axis('off')
 plt.show()
```
